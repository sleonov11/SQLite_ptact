## Задание 1
Для каждого артиста вывести все жанры, которые есть в его песнях, и для каждого жанра вывести наиболее продолжительную песню этого артиста.

```python
%%sql
WITH ArtistGenres AS
(
    SELECT DISTINCT ar.ArtistId, g.GenreId
    FROM artists ar
    JOIN albums al ON ar.ArtistId = al.ArtistId
    JOIN tracks t ON al.AlbumId = t.AlbumId
    JOIN genres g ON t.GenreId = g.GenreId
),
BestTracks AS
(
    SELECT ag.ArtistId, ag.GenreId,
        (
            SELECT t.TrackId
            FROM tracks t
            JOIN albums al ON t.AlbumId = al.AlbumId
            WHERE al.ArtistId = ag.ArtistId AND t.GenreId = ag.GenreId
            ORDER BY t.Milliseconds DESC, t.TrackId ASC
            LIMIT 1
        ) AS TrackId
    FROM ArtistGenres ag
)
SELECT
    ar.Name AS Artist,
    g.Name AS Genre,
    t.Name AS Track,
    t.Milliseconds
FROM BestTracks bt
JOIN artists ar ON bt.ArtistId = ar.ArtistId
JOIN genres g ON bt.GenreId = g.GenreId
JOIN tracks t ON bt.TrackId = t.TrackId
ORDER BY ar.Name, g.Name;
```

## Задание 2
Посчитать скользящее среднее выручки по месяцам с шириной окна 3.


```python
%%sql
WITH monthly_revenue AS
(
    SELECT
        CAST(strftime('%Y', InvoiceDate) AS INTEGER) AS year,
        CAST(strftime('%m', InvoiceDate) AS INTEGER) AS month,
        SUM(Total) AS total
    FROM invoices
    GROUP BY Year, Month
),
monthly_with_index AS
(
    SELECT Year, Month, Total, (Year * 12 + Month) AS MonthIndex
    FROM monthly_revenue
)
SELECT
    m1.Year, m1.Month, m1.Total AS MonthlyTotal,
    (
        SELECT AVG(m2.Total)
        FROM monthly_with_index m2
        WHERE m2.MonthIndex BETWEEN m1.MonthIndex - 2 AND m1.MonthIndex
    ) AS MovingAvg
FROM monthly_with_index m1
ORDER BY m1.Year, m1.Month;
```

## Задание 3
Составить рейтинг клиентов по количеству потраченных ими денег.

```python
%%sql
SELECT
    c.CustomerId,
    c.FirstName || ' ' || c.LastName AS CustomerName,
    SUM(i.Total) AS TotalSpent
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
GROUP BY c.CustomerId
ORDER BY TotalSpent DESC;
```

## Задание 4

Для каждого жанра вывести треки, которые встречались в большем числе заказов,
сумму, потраченную на данные треки, количество купленных треков и компанию, которая больше всех приобрела данную композицию.

```python
%%sql
WITH track_order_stats AS
(
    SELECT t.TrackId, t.GenreId, t.Name AS TrackName,
        COUNT(DISTINCT i.InvoiceId) AS OrderCount,
        SUM(ii.Quantity * t.UnitPrice) AS TotalSpent,
        SUM(ii.Quantity) AS TotalQuantity
    FROM tracks t
    JOIN invoice_items ii ON t.TrackId = ii.TrackId
    JOIN invoices i ON ii.InvoiceId = i.InvoiceId
    GROUP BY t.TrackId, t.GenreId, t.Name
),
genre_top_tracks AS
(
    SELECT tos.*,
        RANK() OVER (PARTITION BY tos.GenreId ORDER BY tos.OrderCount DESC, tos.TotalSpent DESC) as genre_rank
    FROM track_order_stats tos
),
company_track_purchases AS
(
    SELECT ii.TrackId, c.Company,
        SUM(ii.Quantity) AS QuantityBought,
        SUM(ii.Quantity * t.UnitPrice) AS AmountSpent
    FROM invoice_items ii
    JOIN invoices i ON ii.InvoiceId = i.InvoiceId
    JOIN customers c ON i.CustomerId = c.CustomerId
    JOIN tracks t ON ii.TrackId = t.TrackId
    WHERE c.Company IS NOT NULL AND c.Company != ''
    GROUP BY ii.TrackId, c.Company
),
top_company_per_track AS
(
    SELECT ctp.TrackId, ctp.Company,
        ROW_NUMBER() OVER (PARTITION BY ctp.TrackId ORDER BY ctp.AmountSpent DESC, ctp.QuantityBought DESC) as rn
    FROM company_track_purchases AS
 ctp
)
SELECT
    g.Name AS Genre,
    gtt.TrackName AS Track,
    gtt.OrderCount AS OrderCount,
    ROUND(gtt.TotalSpent, 2) AS TotalSpent,
    gtt.TotalQuantity AS TotalQuantity,
    COALESCE(tct.Company, 'No Company') AS TopCompany
FROM genre_top_tracks gtt
JOIN genres g ON gtt.GenreId = g.GenreId
LEFT JOIN top_company_per_track tct ON gtt.TrackId = tct.TrackId AND tct.rn = 1
WHERE gtt.genre_rank = 1
ORDER BY g.Name, gtt.OrderCount DESC, gtt.TotalSpent DESC, gtt.TrackName;
```

## Задание 5

Для каждого плейлиста вывести трек с наибольшей продолжительностью, трек с наименьшей стоимостью (если у треков одинаковая стоимость, то вывести тот, который весит меньше всех), и жанр данного плейлиста (жанром плейлиста будет называться тот жанр, треков которого больше всех в плейлисте).

```python
%%sql
WITH playlist_longest_track AS
(
    SELECT pt.PlaylistId,
        (
            SELECT t.TrackId
            FROM playlist_track pt2
            JOIN tracks t ON pt2.TrackId = t.TrackId
            WHERE pt2.PlaylistId = pt.PlaylistId
            ORDER BY t.Milliseconds DESC, t.TrackId ASC
            LIMIT 1
        ) AS LongestTrackId
    FROM playlist_track pt
    GROUP BY pt.PlaylistId
),
playlist_cheapest_track AS
(
    SELECT pt.PlaylistId,
        (
            SELECT t.TrackId
            FROM playlist_track pt2
            JOIN tracks t ON pt2.TrackId = t.TrackId
            WHERE pt2.PlaylistId = pt.PlaylistId
            ORDER BY t.UnitPrice ASC, t.Bytes ASC, t.TrackId ASC
            LIMIT 1
        ) AS CheapestTrackId
    FROM playlist_track pt
    GROUP BY pt.PlaylistId
),
playlist_top_genre AS
(
    SELECT pt.PlaylistId,
        (
            SELECT t.GenreId
            FROM playlist_track pt2
            JOIN tracks t ON pt2.TrackId = t.TrackId
            WHERE pt2.PlaylistId = pt.PlaylistId
            GROUP BY t.GenreId
            ORDER BY COUNT(*) DESC, t.GenreId ASC
            LIMIT 1
        ) AS TopGenreId
    FROM playlist_track pt
    GROUP BY pt.PlaylistId
)
SELECT p.PlaylistId, p.Name AS PlaylistName, lt.Name AS LongestTrack,
    ct.Name AS CheapestTrack, g.Name AS TopGenre
FROM playlists p
JOIN playlist_longest_track plt ON p.PlaylistId = plt.PlaylistId
JOIN tracks lt ON plt.LongestTrackId = lt.TrackId
JOIN playlist_cheapest_track pct ON p.PlaylistId = pct.PlaylistId
JOIN tracks ct ON pct.CheapestTrackId = ct.TrackId
JOIN playlist_top_genre ptg ON p.PlaylistId = ptg.PlaylistId
JOIN genres g ON ptg.TopGenreId = g.GenreId;
```
